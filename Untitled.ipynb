```python
# IMPORTS
import pandas as pd
import numpy as np

# LOAD DATASET
df = pd.read_csv("air mage.csv")

# BUILD DFS of SKILLS
dfs = []
skills = []
levels = []
for index, row in df.iterrows():
    if pd.isnull(row['Level']) and index != 0:
        dfs.append(pd.DataFrame({'Skill': skills, 'Level': levels}))
        skills = []
        levels = []
    elif index == 0:
        continue
    else:
        skills.append(row['Skill'])
        levels.append(row['Level'])
dfs.append(pd.DataFrame({'Skill': skills, 'Level': levels}))


# DISTRIBUTION
if len(dfs) == 4:
    distribution = [.4, .3, .2, .1]
elif len(dfs) == 3:
    distribution = [.5, .3, .2]
else:
    distribution = [.67, .33]

# START LOOP
levels = [1, 4, 9, 13, 16]
results = {1: [], 4: [], 9: [], 13: [], 16: []}
for level in levels:
    # PART 1
    total_skills = []
    for i in range(len(dfs)):
        # 2 lines to remove incarnate at level 9+ and crippling blow
        if level >= 9:
            dfs[i] = dfs[i].loc[~dfs[i]['Skill'].str.contains('ncarnate')]
            dfs[i] = dfs[i].loc[~dfs[i]['Skill'].str.contains('rippling blow')]            
        skills_at_level = dfs[i].loc[dfs[i]['Level'] <= level]['Skill'].count()
        total_skills.append(skills_at_level)
    total_skills

    # PART 2
    index_order = []
    total = [0]*len(dfs)
    for i in range(sum(total_skills)):
        minimum_score = 100
        minimum_index = None
        minimum_distribution = []
        for i in range(len(dfs)):
            total_temp = total.copy()
            if total_temp[i] + 1 > total_skills[i]:            
                continue
            total_temp[i] += 1
            sum_ = sum(total_temp)
            distribution_temp = []
            for j in range(len(dfs)):
                distribution_temp.append(total_temp[j]/sum_)
            score = 0
            for j in range(len(dfs)):
                score += abs(distribution_temp[j] - distribution[j])
            if score < minimum_score:
                minimum_score = score
                minimum_index = i
                minimum_distribution = distribution_temp
        # print(minimum_score, minimum_index, )
        # print(minimum_distribution)
        index_order.append(minimum_index)
        total[minimum_index] += 1


    # PART 3
    df0_index = 0
    df1_index = 0
    df2_index = 0
    df3_index = 0
    for index in index_order:
        if index == 0:
            while True:
                if dfs[0].iloc[df0_index]['Level'] <= level:
                    results[level].append(dfs[0].iloc[df0_index]['Skill'])
                    df0_index += 1
                    break
                else:
                    df0_index += 1
        elif index == 1:
            while True:
                if dfs[1].iloc[df1_index]['Level'] <= level:
                    results[level].append(dfs[1].iloc[df1_index]['Skill'])
                    df1_index += 1
                    break
                else:
                    df1_index += 1
        elif index == 2:
            while True:
                if dfs[2].iloc[df2_index]['Level'] <= level:
                    results[level].append(dfs[2].iloc[df2_index]['Skill'])                    
                    df2_index += 1
                    break
                else:
                    df2_index += 1
        elif index == 3:
            while True:
                if dfs[3].iloc[df3_index]['Level'] <= level:
                    results[level].append(dfs[3].iloc[df3_index]['Skill'])                    
                    df3_index += 1
                    break
                else:
                    df3_index += 1                
# RESULTS
for k, v in results.items():
    print("Level", k, "skill order")
    print(", ".join(v))
```
